```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread("chess.jpg")
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat_chess, cmap="gray")
```

```python
real_chess = cv2.imread("real_chess.jpg")
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess, cmap="gray")
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
flat_chess[dst > 0.01*dst.max()] = [255, 0, 0]
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
real_chess[dst > 0.01*dst.max()] = [255, 0, 0]
```

```python
plt.imshow(real_chess)
```

## SHI- TOMASI DETECTOR

```python
corners = cv2.goodFeaturesToTrack(image=gray_flat_chess, maxCorners=5, qualityLevel=0.01, minDistance=10) ## maxCorners=-1 return all corners
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess, (x,y), 10, (255, 0, 0 ), -1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(image=gray_flat_chess, maxCorners=80,
                                  qualityLevel=0.01, minDistance=10)  # maxCorners=-1 return all corners

corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess, (x,y), 10, (0, 255, 0), -1)
    
```

```python
plt.imshow(flat_chess)
```
