```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("hachi_face.jpg")
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=150, threshold2= 150)
plt.imshow(edges)
```

```python
# to find the best value for the threshold
med_val = np.median(img)
med_val
```

```python
# lOWER THRESHOLD EITHER 0 OR 70 % 
lower = int(max(0, 0.70*med_val))
upper = int(max(255, 1.3*med_val))
lower, upper
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize=(4,4))
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```
