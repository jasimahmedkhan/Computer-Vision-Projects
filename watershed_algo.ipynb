```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap="gray"):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap=cmap)
```

```python
coins = cv2.imread("pennies_1.jpg")
display(coins)
```

```python
# Media Blur 
# Grayscale
# Binary threshold
# Find the Contours
```

```python
blur = cv2.medianBlur(coins, 25)
display(blur)
```

```python
gray_coins = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)
display(gray_coins)
```

```python
ret, sep_thresh = cv2.threshold(gray_coins, 185, 255, cv2.THRESH_BINARY_INV)
display(sep_thresh)
```

```python
contours, hierarchy = cv2.findContours(sep_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
hierarchy
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(coins, contours, i, (255, 0, 0), 10)
```
